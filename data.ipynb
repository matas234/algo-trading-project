```python
import numpy as np
import pandas as pd
from datetime import datetime
import math

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
data = pd.read_html(url)

sp500_initial = data[0]['Symbol'].to_numpy()
```

```python
current_date = datetime(2024,8,9)

initial_data = {
    'Date': current_date,
    'Ticker List': [sp500_initial]
}

sp500_history = pd.DataFrame(initial_data)
```

```python
changes_history = data[1]

changes_dates = changes_history["Date"]["Date"].to_numpy()
date_format = "%B %d, %Y"
changes_dates = pd.to_datetime(changes_dates, format=date_format).to_numpy()

changes_addedTicker = changes_history["Added"]["Ticker"]
test = changes_addedTicker.isna()
print(changes_addedTicker)


changes_removedTicker = changes_history["Removed"]["Ticker"].to_numpy()
```

```python
array = [[0,2]]

for i in range(1, len(changes_dates)):
    
    if changes_dates[i-1] == changes_dates[i]:

        constant = np.sum(array[-1])
        
        if i < constant :
            continue

        elif i == constant:
            array[-1][1] += 1

        else:
            array.append( [i-1,2] )

print(array)
```

```python
for i in array:
    baseIndex = i[0]
    num = i[1]
    print(i)

    tempArr = changes_addedTicker[baseIndex:baseIndex+num]
    print(tempArr)
```
