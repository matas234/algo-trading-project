```python
import numpy as np
import pandas as pd
from datetime import datetime
import math

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
data = pd.read_html(url)

sp500_initial = data[0]['Symbol'].to_list()
```

```python
changes_history = data[1]

changes_dates = changes_history["Date"]["Date"]
changes_dates = pd.to_datetime(changes_dates, format="%B %d, %Y").to_numpy()

changes_addedTicker = changes_history["Added"]["Ticker"].fillna("").replace({'CDAY': 'DAY', 'FLT':'CPAY', 'Q':'IQV', 'RE':'EG', 'WLTW':'WTW', "DISCK":""}).to_list()

changes_removedTicker = changes_history["Removed"]["Ticker"].replace({'CDAY': 'DAY', 'FLT':'CPAY', 'Q':'IQV', 'RE':'EG','WLTW':'WTW', "DISCK":""}).fillna("").to_list()


print(changes_addedTicker)
```

```python
k=0
length = len(changes_dates)
changes_addedTicker_new = []
changes_removedTicker_new = []
changes_dates_new = []


while k<length:
    temp = np.sum(changes_dates == changes_dates[k])

    changes_addedTicker_new.append(changes_addedTicker[k:k+temp])
    changes_removedTicker_new.append(changes_removedTicker[k:k+temp])
    changes_dates_new.append(changes_dates[k])
    
    k += temp
```

```python
data = {"Date": changes_dates_new,
        "Added": changes_addedTicker_new,
        "Removed": changes_removedTicker_new
        }

data = pd.DataFrame(data)
```
